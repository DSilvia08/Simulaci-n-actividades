```python
#inciso a
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def incisoa(n=100000,a=5,b=10,c=15):
 lista=[]
 for i in range(n):
  u=rnd.random()
  if 0<=u<(b-a)/(c-a):
   x=((u*(c-a)*(b-a)+a**2)**(1/2))+a
   lista.append(x)
  if (b-a)/(c-a) <= u <=1:
   x=((u*(c-a)*(b-c)+c**2)**(1/2))+c
   lista.append(x)
 return lista
```

```python
D=incisoa()
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(D)
```
