```python
#inciso a
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def incisoa(n=1000000,a=5,b=10,c=15):
 lista=[]
 for i in range(n):
  u=rnd.random()
  if 0<=u<(b-a)/(c-a):
   x=np.sqrt(u*(c-a)*(b-a))+a
   lista.append(x)
  else:
   x=c-np.sqrt(c**2-b*c-a*c+a*b+u*(c-a)*(b-c))
   lista.append(x)
 return lista
```

```python
D=incisoa()
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(D)
```

```python
plt.hist(D,bins=50)
```
