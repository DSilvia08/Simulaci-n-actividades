```python
#Importamos las librerías a ocupar
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
```

```python
print(np.sqrt(np.arctan(1)))
```

```python
x=np.linspace(0,1,100) #rango de eje de las x
u_1,u_2=np.random.rand(2) #Generar 2 valores aleatorios U(0,1)
plt.plot(x,np.sqrt(np.arctan(x))) #f(x)
plt.scatter(u_1,u_2,color="red") #Grafica un punto, ya sea fuera o dentro de lafunción
plt.axvline(1,color="pink", linestyle="--")
plt.axhline(np.sqrt(np.arctan(1)),color="pink", linestyle="--")
plt.show()
```

```python
def estimación(n): #ESta función retorna varias listas
 lista1=[] #Estimación de pi
 x_1=[] #Valores para x,y en caso de éxito
 y_1=[]
 x_2=[] #Valores para x,y en caso contrario
 y_2=[]
 éxitos=0 #contador para el número de éxitos
 for i in range(n):
  u_1,u_2=np.random.rand(2) #Genera dos valores aleatorios entre 0 y 1
  u_2=u_2*np.sqrt(np.arctan(1))
  if u_2<=np.sqrt(np.arctan(1)):
   éxitos+=1 #éxito
   x_1.append(u_1)
   y_1.append(u_2)
  else:
   x_2.append(u_1)
   y_2.append(u_2)
  prob=éxitos/(i+1) #probabilidad de pi/4
  lista1.append(4*prob)
 return lista1,x_1,y_1,x_2,y_2
```

```python
G,x_1,y_1,x_2,y_2=estimación(1000)
```

```python
x=np.linspace(0,1,1000) #Rango del eje de las x
plt.plot(x,np.sqrt(np.arctan(x)),color="black")
plt.scatter(x_1,y_1,color="blue") #Éxitos
plt.scatter(x_2,y_2,color="r")
plt.axhline(np.sqrt(np.arctan(1)),color="g", linestyle="--")
plt.show()
```
