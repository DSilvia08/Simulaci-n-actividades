```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unif12(n):
 lista=[]
 for i in range (n) :
  u=0
  for j in range(12) :
   u+=rnd.random()
  x=u-6
  lista.append(x)
 return lista
```

```python
D=unif12(100000)
```

```python
plt.hist(D,bins=50)
```

```python
plt.plot(D)
```

```python
def cdf(muestra):
 n=len(muestra)
 x=np.sort(muestra)
 y=np.arange(0, n) /n
 return x,y
```

```python
x,y=cdf(D)
```

```python
plt.step(x,y)
```
