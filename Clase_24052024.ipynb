```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ejercicio25(n):
  lista=[]
  for i in range(n):
    u=rnd.random()
    x=np.log(u*(np.exp(1)-1)+1)
    lista.append(x)
  return lista
```

```python
S=ejercicio25(1000000)
```

```python
plt.hist(S,bins=50)
```

```python
plt.plot(S)
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(S)
```

```python
plt.step(x,y)
```
