```python
def A(n):
  contador=0
  a=[]
  b=[]
  for i in range(n):
    u_1,u_2=np.random.rand(2)
    if u_2<=np.sqrt(1-u_1**2):
      contador+=1
      a.append(u_1)
      b.append(u_2)
  C=(4*(contador/n))
  return C,a,b
```
