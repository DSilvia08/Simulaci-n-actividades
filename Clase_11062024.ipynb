#Estimación del número pi

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
```

```python
x=np.linspace(0,1,100)
u_1,u_2=np.random.rand(2)
plt.plot(x,np.sqrt(1-x**2))
plt.scatter(u_1,u_2,color="red")
plt.show()
```

```python
def A(n):
  contador=0
  a=[]
  b=[]
  for i in range(n):
    u_1,u_2=np.random.rand(2)
    if u_2<=np.sqrt(1-u_1**2):
      contador+=1
    a.append(u_1)
    b.append(u_2)
  C=4*(contador/n)
  return C,a,b
```

```python
E,a,b=A(100)
print(E)
```

```python
x=np.linspace(0,1,100)
plt.plot(x,np.sqrt(1-x**2))
plt.scatter(a,b,color="red")
plt.show()
```

```python
def B(n):
  c=[]
  d=[]
  for i in range(n):
    Z,a,b=A(n)
  c.append(a)
  d.append(b)
  return Z,c,d
```

```python
H,c,d=B(20)
print(H)
```

```python
x=np.linspace(0,1,100)
plt.plot(x,np.sqrt(1-x**2))
plt.scatter(c,d,color="red")
plt.show()
```
