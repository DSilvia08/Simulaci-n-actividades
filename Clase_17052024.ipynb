```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(n):
  lista1=[]
  for i in range(n):
    R_1=rnd.random()
    R_2=rnd.random()
    if 0<R_1<(5-2)/(8-2):
      x=2+(8-2)*np.sqrt(R_2)
      lista1.append(x)
    else:
      x=8+(8-5)*np.sqrt(1-R_2)
      lista1.append(x)
  return lista1
```

```python
n=1000000
S=f(n)
```

```python
plt.hist(S,bins=50)
```

```python
plt.plot(S)
```

```python
def cdf(G):
 n=len(G)
 x=np.sort(G)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(S)
```

```python
plt.step(x,y)
```
