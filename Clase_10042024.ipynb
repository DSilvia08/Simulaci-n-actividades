```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def uniform(a=15,b=10,n=10):
 lista=[]
 for i in range(n):
  u=rnd.random()
  x=((b-a)*u)+a
  lista.append(x)
 return lista
```

```python
D=uniform(15,10,1000)
```

```python
mu=np.mean(D)
print(mu)
```

```python
sigma=np.var(D)
print(sigma)
```

```python
plt.hist(D)
```

```python
plt.plot(D)
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(D)
```

```python
plt.step(x,y)
plt.axvline(mu, color='r', linestyle='--', label='media')
plt.axvline(mu + np.sqrt(sigma), color='g', linestyle='--', label='Media + Desviacin Estndar')
plt.axvline(mu - np.sqrt(sigma), color='g', linestyle='--', label='Media - Desviacin Estndar')

plt.legend()
plt.show()
```
