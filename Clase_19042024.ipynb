```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

#Ejercicio 1

```python
def ejercicio1(n):
 lista=[]
 for i in range(n):
  u=rnd.random()
  if 0<=u<1/4:
   x=4*u
   lista.append(x)
  else:
   x=((19/2)+((-39/4 + 20*u)**1/2))/10
   lista.append(x)
 return lista
```

```python
D=ejercicio1(1000000)
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(D)
```

```python
plt.hist(D)
```

```python
plt.plot(D)
```

```python
plt.step(x,y)
```
