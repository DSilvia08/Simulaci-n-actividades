```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def incisob(n):
 lista=[]
 for i in range(n):
  u=rnd.random()
  if 0<=u<3/4:
   x=(5/4)-((25/16 -2*u)**(1/2))
   lista.append(x)
  else:
   x=(4*u)+1
   lista.append(x)
 return lista
```

```python
D=incisob(10000)
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(D)
```
