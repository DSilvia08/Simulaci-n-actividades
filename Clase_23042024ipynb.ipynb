```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ejemplo(n):
 lista=[]
 for i in range(n):
  u=rnd.random()
  x=((u**0.135)-(1-u)**0.135)/0.1975
  lista.append(x)
 return lista
```

```python
D=ejemplo(10000)
```

```python
def cdf(D):
 n=len(D)
 x=np.sort(D)
 y=np.arange(0,n)/n
 return x,y
```

```python
x,y=cdf(D)
```

```python
plt.hist(D,bins=50)
```

```python
plt.plot(D)
```

```python
plt.step(x,y)
```
