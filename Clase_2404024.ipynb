```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inversa (n, p):
 lista=[]
 for i in range(n) :
  u=rnd.random()
  if u>= 1-p:
   x=1
   lista.append(x)
  else:
   x=0
   lista.append(x)
 return lista
```

```python
D=inversa (1000, 0.35)
```

```python
def cdf(muestra):
 n=len(muestra)
 x=np.sort(muestra)
 y=np.arange(0, n) /n
 return x,y
```

```python
x,y =cdf(D)
```

```python
plt.hist(D, bins=50)
```

```python
plt.plot(D)
```
